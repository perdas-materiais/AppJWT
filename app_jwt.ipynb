```python
import jwt
from datetime import datetime, timedelta, timezone
```

### ‚úÖ CRIACAO DO TOKEN

```python
# CRIACAO DO TOKEN
SECRET_KEY = "minha_chave_secreta"
ALGORITHM = "HS256"

payload = {
    "user_id": 1,
    "username": "renato",
    "email": "renatolcorrea@gmail.com",
    "dpto":"transmissao",
    "data_nascimento":"1965-04-16",
    "exp": datetime.now(timezone.utc) + timedelta(minutes=30)
}

token = jwt.encode(payload, SECRET_KEY, algorithm=ALGORITHM)

print(f"Data hora atual: {datetime.now(timezone.utc)}")
print(f"VALOR DO TOKEN CRIADO: {token}")
```

### ‚úÖ DECODIFICANDO O TOKEN

```python
try:
    decoded = jwt.decode(
        token,
        SECRET_KEY,
        algorithms=[ALGORITHM]
    )
    print(f"üéñÔ∏è VALOR TOTAL DO TOKEN:\n{decoded}")
    print(f"\nüéØ Recuperando dados principais:\n")
    print(f"üìå Username: {decoded["username"]}")
    print(f"üìå Email: {decoded["email"]}")
    print(f"üìå Dpto: {decoded["dpto"]}")
    print(f"üìå Data Nascimento: {decoded["data_nascimento"]}")
    print(f"üìå Expricao epoch: {decoded["exp"]}, Data Hora: {datetime.fromtimestamp(decoded["exp"])}")

except jwt.ExpiredSignatureError:
    print(f"üö® Token expirado")
except jwt.InvalidTokenError:
    print(f"üö® Token inv√°lido")
```
